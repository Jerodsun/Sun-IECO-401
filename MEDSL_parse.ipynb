```python
df = pd.read_csv("data/MEDSL/2012-2018-house.csv")
```

```python
df = df[["year", "state_po", "district", "candidate", "party", "candidatevotes", "totalvotes", "fusion_ticket"]]
```

```python
len(df)
```

```python
df.head()
```

```python
df_fusion = df[df["fusion_ticket"] == 1]
df_fusion.tail()
```

```python
df_fusion = df_fusion.sort_values(by=["year", "state_po", "district", "candidatevotes"], ascending=[1,1,1,0])
df_fusion
```

```python
fusion_total = df_fusion.groupby(["year", "state_po", "district", "candidate"])["candidatevotes"].sum().reset_index()
fusion_total
```

```python
df_fusion.sort_values(by=["year", "state_po", "district", "candidatevotes"], ascending=[1,1,1,0], inplace=True)
df_fusion
```

```python
result = df_fusion.merge(fusion_total, on=["year", "state_po", "district", "candidate"])
result
```

```python
result.sort_values(by=["year", "state_po", "district", "candidatevotes_y"], ascending=[1,1,1,0], inplace=True)
result = result.drop_duplicates(subset=["year", "state_po", "district", "candidate"]).drop(columns=['candidatevotes_x'])
result = result.rename({"candidatevotes_y": "candidatevotes"}, axis=1)
```

```python
result
```

```python
df_orig = df[df["fusion_ticket"] == 0]
house_fused = df_orig.append(result)
```

```python
house_fused
```

```python
house_fused.to_csv("2012-2018_house_fused.csv", index=False)
```
